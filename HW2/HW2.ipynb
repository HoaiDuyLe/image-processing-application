```python
# import the image module
from PIL import Image 

# Read a color image
# grayImage = Image.open("./gray.jpeg")
grayImage = Image.open("./rose.jpg")
# grayImage = Image.open("./monalisa.jpg")
```

```python
import numpy as np
import matplotlib.pyplot as plt
fig, axes = plt.subplots(ncols=2, nrows=2, figsize=(12, 12), gridspec_kw = {'wspace':0, 'hspace':0.1})
np.vectorize(lambda ax:ax.axis('off'))(axes)
ax0, ax1, ax2, ax3= axes.flat

ax0.imshow(grayImage, cmap=plt.cm.gray)
ax0.set_title('24-bit', fontsize=18)

image20bit = grayImage.convert("P", palette=Image.ADAPTIVE, colors=20)
ax1.imshow(image20bit, cmap=plt.cm.gray)
ax1.set_title('20-bit', fontsize=18)

image16bit = grayImage.convert("P", palette=Image.ADAPTIVE, colors=16)
ax2.imshow(image16bit, cmap=plt.cm.gray)
ax2.set_title('16-bit', fontsize=18)

image8bit = grayImage.convert("P", palette=Image.ADAPTIVE, colors=8)
ax3.imshow(image8bit, cmap=plt.cm.gray)
ax3.set_title('8-bit', fontsize=18)

# Save result.
fig.savefig("D:/result_HW2_1.png",bbox_inches='tight')
```

```python
import PIL
grayImage = Image.open("./rose.jpg")
width, height = grayImage.size

grayImage1 = grayImage.resize((int(width/2), int(height/2)))
grayImage1.show()
grayImage2 = grayImage.resize((int(width/4), int(height/4)))
grayImage2.show()
grayImage3 = grayImage.resize((int(width/8), int(height/8)))
grayImage3.show()
grayImage4 = grayImage.resize((int(width/16), int(height/16)))
grayImage4.show()
grayImage5 = grayImage.resize((int(width/32), int(height/32)))
grayImage5.show()
```

```python
grayImage1_duplicate = grayImage1.resize((width, height),PIL.Image.LANCZOS)
grayImage1_duplicate.show()
grayImage2_duplicate = grayImage2.resize((width, height))
grayImage2_duplicate.show()
grayImage3_duplicate = grayImage3.resize((width, height))
grayImage3_duplicate.show()
grayImage4_duplicate = grayImage4.resize((width, height))
grayImage4_duplicate.show()
grayImage5_duplicate = grayImage5.resize((width, height),PIL.Image.LANCZOS)
grayImage5_duplicate.show()
```

```python

grayImage3_nearest = grayImage3.resize((width, height),PIL.Image.NEAREST)
grayImage3_nearest.show()
grayImage4_nearest = grayImage4.resize((width, height),PIL.Image.NEAREST)
grayImage4_nearest.show()
grayImage5_nearest = grayImage5.resize((width, height),PIL.Image.NEAREST)
grayImage5_nearest.show()
```

```python
grayImage3_bilinear = grayImage3.resize((width, height),PIL.Image.BILINEAR )
grayImage3_bilinear.show()
grayImage4_bilinear = grayImage4.resize((width, height),PIL.Image.BILINEAR )
grayImage4_bilinear.show()
grayImage5_bilinear = grayImage5.resize((width, height),PIL.Image.BILINEAR )
grayImage5_bilinear.show()
```

```python
fig, axes = plt.subplots(ncols=2, nrows=3, figsize=(12, 18), gridspec_kw = {'wspace':0, 'hspace':0.01})
np.vectorize(lambda ax:ax.axis('off'))(axes)
ax0, ax1, ax2, ax3, ax4, ax5 = axes.flat

ax0.imshow(grayImage3_nearest, cmap=plt.cm.gray)
ax0.set_title('Nearest neighor interpolation', fontsize=18)

ax2.imshow(grayImage4_nearest, cmap=plt.cm.gray)
# ax1.set_title('4-bit', fontsize=18)

ax4.imshow(grayImage5_nearest, cmap=plt.cm.gray)
# ax2.set_title('6-bit', fontsize=18)

ax1.imshow(grayImage3_bilinear, cmap=plt.cm.gray)
ax1.set_title('Bilinear Interpolation', fontsize=18)

ax3.imshow(grayImage4_bilinear, cmap=plt.cm.gray)
# ax4.set_title('2-bit', fontsize=18)

ax5.imshow(grayImage5_bilinear, cmap=plt.cm.gray)
# ax5.set_title('1-bit', fontsize=18)

# Save result.
fig.savefig("D:/result_HW2_4.png",bbox_inches='tight')
```
